```python
#Label Encoding for object to numeric conversion
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for c1 in a:
    df[c1] = le.fit_transform(df[c1].astype(str))

print (df.info())
```

```python
y=df["h1n1_vaccine"]
# x=df[["h1n1_worry","h1n1_awareness",'dr_recc_h1n1_vacc', 'dr_recc_seasonal_vacc',
#        'chronic_medic_condition', 'cont_child_undr_6_mnths',
#        'is_health_worker', 'has_health_insur', 'is_h1n1_vacc_effective',
#        'is_h1n1_risky', 'sick_from_h1n1_vacc', 'is_seas_vacc_effective',
#        'is_seas_risky']]

x=df.iloc[:,df.columns!="h1n1_vaccine"]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape) 
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression 
from sklearn import metrics

lr= LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred_lr=lr.predict(x_test)



#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix 
from sklearn import metrics

cm= confusion_matrix(y_test,y_pred_lr)  
print("Confusion matrix: \n",cm)

from sklearn.metrics import accuracy_score
print('\nAccuracy: ',accuracy_score(y_test, y_pred_lr))
knn_model=round(100*(accuracy_score(y_test, y_pred_lr)),2)
score = lr.score(x_test, y_test)

plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);


from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred_lr)
roc_auc = auc(false_positive_rate, true_positive_rate)
print("roc_auc",roc_auc)
```

```python
from sklearn.linear_model import SGDClassifier

clf = SGDClassifier(loss='log',alpha=0.001,max_iter=100)
clf.fit(x_train,y_train)

y_pred_sgd = clf.predict(x_test)





#Creating the Confusion matrix  
from sklearn.metrics import confusion_matrix 
from sklearn import metrics

cm= confusion_matrix(y_test,y_pred_sgd)  
print("Confusion matrix: \n",cm)

from sklearn.metrics import accuracy_score
print('\nAccuracy: ',accuracy_score(y_test, y_pred_sgd))
knn_model=round(100*(accuracy_score(y_test, y_pred_sgd)),2)
score = clf.score(x_test, y_test)

plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);


from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred_sgd)
roc_auc = auc(false_positive_rate, true_positive_rate)
print("roc_auc",roc_auc)
```
