# Decision Tree Model(entropy)

```python
x=df.iloc[:,:13]
y=df.iloc[:,13:]
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 1111)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

x_train = sc.fit_transform(x_train) 
x_test = sc.transform(x_test)  
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)

dt.fit(x_train, y_train)
```

```python
y_pred_dt = dt.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(y_test,y_pred_dt)  
print("Confusion matrix: \n",cm)

from sklearn.metrics import accuracy_score
print('\nAccuracy: ',accuracy_score(y_test, y_pred_dt))
score = dt.score(x_test, y_test)

plt.figure(figsize=(9,9))
sns.heatmap(cm, annot=True, fmt=".3f", linewidths=.5, square = True, cmap = 'Blues_r');
plt.ylabel('Actual label');
plt.xlabel('Predicted label');
all_sample_title = 'Accuracy Score: {0}'.format(score)
plt.title(all_sample_title, size = 15);


from sklearn.metrics import roc_curve, auc
false_positive_rate, true_positive_rate, thresholds = roc_curve(y_test, y_pred_dt)
roc_auc = auc(false_positive_rate, true_positive_rate)
print("\nroc_auc: ",roc_auc)
```

# Hypertuning

```python
from sklearn.model_selection import GridSearchCV

pGrid = {'max_depth': range(2, 10),
        'min_samples_leaf': range(10, 51, 10),
        'min_samples_split': range(20, 81, 20),
        'max_features': [5,10,15,20,25]} 

gscv = GridSearchCV(estimator = DecisionTreeClassifier(), param_grid = pGrid, cv = 5,scoring = 'recall', n_jobs = -1, verbose = True)
gscv.fit(x,y)
```
