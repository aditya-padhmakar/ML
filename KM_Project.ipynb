```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df["Gender"] = LE.fit_transform(df["Gender"])

df.head()
```

```python
# Applying PCA function on training and testing set of X component 
from sklearn.decomposition import PCA 
  
pca = PCA(n_components = 2) 
  
x = pca.fit_transform(df) 
# x_test = pca.transform(x_test) 
  
explained_variance = pca.explained_variance_ratio_ 
```

```python
x.shape
```

```python
x
```

```python
# Using the elbow method to find the optimal number of clusters


from sklearn.cluster import KMeans
ssd = []
for num_clusters in list(range(1,15)):
    km = KMeans(n_clusters = num_clusters, max_iter=50)
    km.fit(x)
    ssd.append(km.inertia_)

# plt.plot(ssd)
# plt.show()
plt.plot(ssd)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('SSD')
plt.show()
```

according to elbow method optimal value of k is 2

```python
# The silhouette value measures how similar a point is to its own cluster (cohesion) compared to other clusters (separation).

#1st way to calculate silhouette score

sill=[]
from sklearn.metrics import silhouette_score
for i in range(2,15):
    
    # intialise kmeans
    kmeans = KMeans(n_clusters=i, max_iter=50)
    kmeans.fit(x)
    
    cluster_labels = kmeans.labels_
    
    # silhouette score
    silhouette_avg = silhouette_score(x, cluster_labels)
    sill.append(silhouette_avg)
    print("For n_clusters={0}, the silhouette score is {1}".format(num_clusters, silhouette_avg))
```

```python
#2nd way to check silhouette score

from sklearn.metrics import silhouette_score
sil = []
kmax = 10

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
  kmeans = KMeans(n_clusters = k).fit(x)
  labels = kmeans.labels_
  sil.append(silhouette_score(x, labels, metric = 'euclidean'))
```

```python
plt.plot(sill)
```

```python
plt.plot(sil)
```

according to silhouette score, the optimal number of k is 2

```python
# Fitting K-Means to the dataset

kmeans = KMeans(n_clusters = 2, init = 'k-means++')
y_kmeans = kmeans.fit_predict(x)

# kmeans = KMeans(n_clusters=2, max_iter=50)
# kmeans.fit(x)
# y_kmeans1 = kmeans.labels_
```

```python
plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 100, c = 'red', label = 'Cluster 1')

plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 100, c = 'blue', label = 'Cluster 2')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroids')
plt.legend()
plt.show()
```
